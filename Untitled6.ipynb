# Data Cleaning

```python
data=df=pd.read_csv('Telco-Customer-Churn.csv')
```

```python
data.info()
```

```python
cols=['customerID']
df=data.drop(cols,axis="columns")
df.head()
```

```python
pd.to_numeric(df['TotalCharges'],errors='coerce')
df[pd.to_numeric(df['TotalCharges'], errors='coerce').isnull()].shape
df=df[df['TotalCharges']!= ' ']
```

```python
df.head(10)
```

```python
def print_unique(df):
    for cols in df:
        if df[cols].dtypes=='object':
            print(f'{cols} : {df[cols].unique()}')
print_unique(df)
```

```python
df.replace('No phone service', 'No', inplace=True)
df.replace('No internet service', 'No', inplace=True)
```

```python
print_unique(df)
```

```python
colums=['Partner','Dependents','OnlineBackup','PhoneService','MultipleLines','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for i in colums:
    df[i].replace({'Yes': 1, 'No':0},inplace=True)
```

```python
df.head()
```

```python
df['gender'].replace({'Male':1, 'Female':0},inplace =True)
df.head()
```
